```python
import pandas as pd
import numpy as np
import json
```

```python
table = pd.read_html("https://www.shogi.or.jp/player/numerical_order.html")[0]
```

```python
table
```

```python
player = {}
for i, j  in zip(table["番号"], table["棋士名"]):
    if not np.isnan(i):
        player[j] = int(i)
for i, j in zip(table["番号.1"], table["棋士名.1"]):
    if not np.isnan(i):
        player[j] = int(i)
```

```python
path = "./player_num.json"
json_file = open(path, mode="w+", encoding="utf-8")
json.dump(player, json_file, indent=2, ensure_ascii=False)
json_file.close()
```
